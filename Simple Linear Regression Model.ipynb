```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
salim = pd.read_csv(r"datasets_141319_332156_1.01. Simple linear regression.csv")
salim.head(10)
```

```python
x = salim['SAT'].values
y= salim['GPA'].values
x = x.reshape(-1, 1)
y = y.reshape(-1, 1)
scatter = plt.scatter(x,y)
plt.title('SCATTER PLOT FOR THE DATASET')
plt.xlabel('GPA OF THE STUDENTS')
plt.ylabel('SAT SCORE OF THE STUDENTS')
```

```python
#Evaluating the correlation between both variables
plt.title('CORRELATION HEATMAP BETWEEN THE SAT SCORES AND GPA OF THE STUDENTS')
sns.set_style("white")
sns.set_palette("husl")
sns.heatmap(salim.corr(), square=True, annot=True)
plt.show()
```

```python
#Visualizing the line of best fit for the linear regression

sns.lmplot('SAT', 'GPA',salim).savefig('CORRELATION HEATMAP BETWEEN THE SAT SCORES AND GPA OF THE STUDENTS')
plt.title('CORRELATION HEATMAP BETWEEN THE SAT SCORES AND GPA OF THE STUDENTS')
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = 0.25)
```

```python
from sklearn.linear_model import LinearRegression
regressor= LinearRegression()
regressor.fit(x_train, y_train)
y_pred = regressor.predict(x_test)
print(y_pred)
```

```python
from sklearn.metrics import r2_score
accu = r2_score(y_test, y_pred)
```

```python
print('The Accuracy Score of the model is given as '+ (str(accu*100)) +'%. The low accuracy is due to the precision of the model - To 8 decimal places' )
```
